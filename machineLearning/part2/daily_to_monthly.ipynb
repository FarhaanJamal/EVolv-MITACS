```python
import pandas as pd

data = pd.read_csv("data_cleaning/all-users-daily-data.csv")
data.head(50)
```

```python
print(f"Tot records: {data.shape}")
print(data.dtypes)
data.nunique()
```

```python
data.isnull().sum()
```

```python
rows = []

start_date = ""
initial_energy = 0
final_energy = 0
user_id = ""

#count = 0
for ind, val in data.iterrows():
    if str(val["date"])[-2:] == "01":
        if data.iloc[ind-1]["userId"] == user_id:
            final_energy = data.iloc[ind-1]["energy"]
            monthly_consumption = (final_energy - initial_energy) / (pow(10, 10)) * 3
            new_row = {"userId" : user_id, 'date': start_date, 'energy': monthly_consumption}
            rows.append(new_row)
        user_id = val["userId"]
        start_date = val["date"]
        initial_energy = val["energy"]
    """count += 1
    if count == 100: break"""

last_date = str(data.iloc[-1]["date"])[-2:]
if last_date in ["28", "29", "30", "31"]:
    if data.iloc[-1]["userId"] == user_id:
        monthly_consumption = (data.iloc[-1]["energy"] - initial_energy) / (pow(10, 10))
        new_row = {"userId" : user_id, 'date': start_date, 'energy': monthly_consumption}
        rows.append(new_row)
```

```python
column_names = data.columns

df = pd.DataFrame(rows, columns=column_names)
df.head()
```

```python
print(f"Tot records: {df.shape}")
print(df.dtypes)
df.nunique()
```

```python

df.to_csv("data_cleaning/all-users-monthly-data.csv", index=False, header=True)
```
