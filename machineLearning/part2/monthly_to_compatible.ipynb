```python
import pandas as pd

data = pd.read_csv("data_cleaning/all-users-monthly-data.csv")
data.head(50)
```

```python
print(f"Tot records: {data.shape}")
print(data.dtypes)
data.nunique()
```

```python
rows = []

user_id = ""
sum_energy = 0
month_count = 1

#count = 0
for ind, val in data.iterrows():
    if data.iloc[ind-1]["userId"] == user_id:
        sum_energy += val["energy"]
        month_count += 1
    else:
        new_row = {"userId" : user_id, 'energy': sum_energy / month_count}
        rows.append(new_row)
        user_id = val["userId"]
        month_count = 1
        sum_energy = val["energy"]
    
    """count += 1
    if count == 100: break"""

new_row = {"userId" : user_id, 'energy': sum_energy / month_count}
rows.append(new_row)

rows.pop(0)
```

```python
print(rows)
```

```python
df = pd.DataFrame(rows, columns=['userId', 'energy'])
df.head()
```

```python
print(f"Tot records: {df.shape}")
print(df.dtypes)
df.nunique()
```

```python
df.describe()
```

```python

df.to_csv("data_cleaning/energy-only.csv", index=False, header=True)
```
