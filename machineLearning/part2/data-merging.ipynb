```python
import pandas as pd

data1 = pd.read_csv("data_cleaning/household-information.csv")
pd.set_option('display.max_columns', 9)
pd.set_option('display.max_rows', None)
data1.head()
```

```python
print(f"Tot records: {data1.shape}")
print(data1.dtypes)
```

```python
data2 = pd.read_csv("data_cleaning/energy-only.csv")
data2.head()
```

```python
print(f"Tot records: {data2.shape}")
print(data2.dtypes)
data2.nunique()
```

```python
data = pd.merge(data1, data2, on='userId', how="left")
data.head()
```

```python
print(f"Tot records: {data.shape}")
print(data.dtypes)
```

```python
data.isnull().sum()
```

```python
print(data.describe(include='all'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(data.isnull() , cmap='viridis', ax=ax)
```

```python
print(f"Total null in the dataset: {data.isnull().sum().sum()}")
```

```python
data.drop(columns=['userId', 'heatingTypeOther', 'waterHeaterTypeOther'], inplace=True)
```

```python
for col in data.columns:
    if col not in ['userId', 'energy']:
        plt.figure(figsize=(12, 6))
        plt.title(f'Distribution of values in "{col}" column')
        plt.xlabel(col)
        plt.ylabel('Count')
        value_counts = data[col].value_counts()
        sns.barplot(x=value_counts.index, y=value_counts.values, hue=value_counts.index, palette='viridis')
        if isinstance(value_counts.index[0], str):
            plt.xticks(rotation=90)
        plt.show()
```

```python
print(data.columns)
```

```python
data = data.drop(columns=['heatingTypes', 'waterHeaterTypes', 'coolingType', 'HEAT_PUMP', 'FRIDGE_COMBO', 'POOL_PUMP', 'FLOW_WATER_HEATER', 'TABLET', 'HEATING_FAN', 'DVD', 'OTHERS'])
```

```python
data['numberOfPeople'] = data['numberOfPeople'].replace('8+', '8')
data['numberOfChildren'] = data['numberOfChildren'].replace('7+', '7')

data["numberOfPeople"] = pd.to_numeric(data["numberOfPeople"])
data["numberOfChildren"] = pd.to_numeric(data['numberOfChildren'])
```

```python

cols = ["squareMeters", "energy"]

for col in cols:
    data[col] = data[col].fillna(data[col].mean())
```

```python
cols = ["numberOfPeople", "numberOfChildren", "homeType"]

for col in cols:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
print(f"Total null in the dataset: {data.isnull().sum().sum()}")
```

```python
data['homeType'].replace({'MULTI_FAMILY': 1, 'SINGLE_FAMILY': 0}, inplace=True)
```

```python
data.to_csv("data_cleaning/final-cleaned-monthly-data.csv", index=False, header=True)
data.head()
```
